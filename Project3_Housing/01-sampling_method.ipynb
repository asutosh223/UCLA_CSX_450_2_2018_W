#### Repeat Sampling

```python
sample_means = []
for _ in range(10):
    sample_means.append(housing_df.sample(5).mean())

sample_means = np.array(sample_means)
(sample_means.mean(axis=0)-stats['mean'])/stats['std']
```

```python
sample_means = []
for _ in range(50):
    sample_means.append(housing_df.sample(5).mean())

sample_means = np.array(sample_means)
(sample_means.mean(axis=0)-stats['mean'])/stats['std']
```

```python
sample_means = []
for _ in range(100):
    sample_means.append(housing_df.sample(5).mean())

sample_means = np.array(sample_means)
(sample_means.mean(axis=0)-stats['mean'])/stats['std']
```
